```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Choosing plot style in matplotlib
from matplotlib import style
```

```python
style.use("ggplot")
```

```python
# Read the CSV files into Pandas DataFrames
orders_df = pd.read_csv("orders.csv")
order_details_df = pd.read_csv("order_details.csv")
pizzas_df = pd.read_csv("pizzas.csv")
pizza_types_df = pd.read_csv("pizza_types.csv", encoding="latin1")

# Display the first 5 rows of each DataFrame
print("\nFirst 5 rows of 'Orders':")
display(orders_df.head())

print("\nFirst 5 rows of 'Order Details':")
display(order_details_df.head())

print("\nFirst 5 rows of 'Pizzas':")
display(pizzas_df.head())

print("\nFirst 5 rows of 'Pizza Types':")
display(pizza_types_df.head())

# Display columns and their data types for each DataFrame
print("\nColumn information for 'Orders':")
print(orders_df.info())

print("\nColumn information for 'Order Details':")
print(order_details_df.info())

print("\nColumn information for 'Pizzas':")
print(pizzas_df.info())

print("\nColumn information for 'Pizza Types':")
print(pizza_types_df.info())
```

```python
# Merging order_details_df with orders_df on 'order_id'
merged_df = pd.merge(order_details_df, orders_df, on='order_id', how='outer')

# Display the first 5 rows of the merged DataFrame
print("\nFirst 5 rows after merging 'Order Details' and 'Orders':")
display(merged_df.head())

# Merging the result with pizzas_df on 'pizza_id'
merged_df = pd.merge(merged_df, pizzas_df, on='pizza_id', how='outer')

# Display the first 5 rows of the merged DataFrame
print("\nFirst 5 rows after merging with 'Pizzas':")
display(merged_df.head())

# Merging the result with pizza_types_df on 'pizza_type_id'
final_merged_df = pd.merge(merged_df, pizza_types_df, on='pizza_type_id', how='outer')

# Display the first 5 rows of the merged DataFrame
print("\nFirst 5 rows after merging with 'Pizza Types':")
display(final_merged_df.head())

# Display columns and their data types of the merged DataFrame
print("\nColumn information after merging with 'Pizza Types':")
print(final_merged_df.info())
```

### **Question 1**
1.	What is the total revenue/sales?

```python
# Calculate total revenue by multiplying quantity by price
final_merged_df['revenue'] = final_merged_df['quantity'] * final_merged_df['price']

# Sum the revenue column to get the total revenue
total_revenue = final_merged_df['revenue'].sum()

# Display the total revenue
print(f"The total revenue is: ${total_revenue:,.2f}")
```

### **Question 2**
2.	Find the total quantity sold.

```python
# Calculate the total quantity sold
total_quantity_sold = final_merged_df['quantity'].sum()

# Display the total quantity sold
print(f"The total quantity of pizzas sold is: {total_quantity_sold}")
```

### **Question 3**
3.	Find the total orders.

```python
# Calculate the total number of orders
total_orders = final_merged_df['order_id'].nunique()

# Display the total number of orders
print(f"The total number of orders is: {total_orders:,.0f}")
```

### **Question 4**
4.	How many pizza types do they sell?

```python
# Calculate the number of unique pizza types
num_pizza_types = final_merged_df['pizza_type_id'].nunique()

# Display the number of unique pizza types
print(f"The number of unique pizza types sold is: {num_pizza_types}")
```

### **Question 5**
5.	Find the average price of the pizzas.

```python
# Calculate the average price of the pizzas
average_price = final_merged_df['price'].mean()

# Display the average price of the pizzas
print(f"The average price of the pizzas is: ${average_price:,.2f}")
```

### **Question 6**
6.	What are the peak hours of sales?

```python
# Convert 'time' column to datetime objects with specified format
final_merged_df['time'] = pd.to_datetime(final_merged_df['time'], format='%H:%M:%S')

# Extract hour from the 'time' column
final_merged_df['hour'] = final_merged_df['time'].dt.hour

# Group by hour and count the number of orders
hourly_order_counts = final_merged_df.groupby('hour').size().reset_index(name='order_count')

# Sort by order count to find peak hours
peak_hours = hourly_order_counts.sort_values(by='order_count', ascending=False)

# Display the peak hours
print("Peak hours of sales:\n")
# display(peak_hours)
display(peak_hours.head())
```

The peak hours of sales are between `12:00 PM`and `1:00 PM`.

```python
# Visualizing the Peak hours of sales using bar chart

y = peak_hours.order_count
x = peak_hours.hour

#Setting my figure size
plt.figure(figsize=(12,5))

plt.bar(x, y, color="green")

plt.title('Peak Hours of Sales')
plt.xlabel('Hours')
plt.ylabel('Sales Recorded')

plt.show()
```

### **Question 7**
7.	Find the total sales made on each day of the week. Which day of the week is when sales are made the most?

```python
# Convert 'date' column to datetime
final_merged_df['date'] = pd.to_datetime(final_merged_df['date'])

# Extract day of the week
final_merged_df['day_of_week'] = final_merged_df['date'].dt.day_name()

# Group by day of the week and sum the revenue
daily_sales = final_merged_df.groupby('day_of_week')['revenue'].sum().reset_index(name='total_sales')

# Sort by total sales
most_sales_day = daily_sales.sort_values(by='total_sales', ascending=False)

# Display the total sales for each day of the week
print("Total sales for each day of the week:\n")
display(daily_sales)

# Display the day with the most sales
print(f"\n{most_sales_day.iloc[0]['day_of_week']} is the day with the most sales (${most_sales_day.iloc[0]['total_sales']:.2f})")
```

```python
# Visualizing the day with most sales using bar chart

x = daily_sales.day_of_week
y = daily_sales.total_sales

#Setting my figure size
plt.figure(figsize=(12,5))

plt.bar(x, y, color="green")

plt.title('Day with most Sales')
plt.xlabel('Day of the week')
plt.ylabel('Sales Recorded')

plt.show()
```

### **Question 8**
8.	Find the top 5 bestselling pizzas.

```python
# Group by pizza name and sum the quantity
pizza_sales = final_merged_df.groupby('name')['quantity'].sum().reset_index(name='total_quantity')

# Sort by total quantity sold
top_selling_pizzas = pizza_sales.sort_values(by='total_quantity', ascending=False)

# Display the top 5 best-selling pizzas
print("Top 5 best-selling pizzas:\n")
display(top_selling_pizzas.head())
```

```python
# Visualizing the best selling pizza using bar chart

x = top_selling_pizzas.name
y = top_selling_pizzas.total_quantity

#Setting my figure size
plt.figure(figsize=(10,12))

plt.barh(x, y, color="green")

plt.title('Best Selling Pizza Types')
plt.xlabel('Day of the week')
plt.ylabel('Quantity of Pizza')

plt.show()
```

### **Question 9**
9.	Find the sales made in each month. Any trend noticeable?

```python
import pandas as pd

# Extract month from the 'date' column
final_merged_df['month'] = final_merged_df['date'].dt.month_name()

# Group by month and sum the revenue
monthly_sales = final_merged_df.groupby('month')['revenue'].sum().reset_index()

# Order the months correctly
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_sales['month'] = pd.Categorical(monthly_sales['month'], categories=month_order, ordered=True)
monthly_sales = monthly_sales.sort_values('month')

# Display the sales for each month
print("Sales for each month:\n")
display(monthly_sales)

# Analyze trends (basic trend observation)
highest_sales_month = monthly_sales.loc[monthly_sales['revenue'].idxmax(), 'month']
lowest_sales_month = monthly_sales.loc[monthly_sales['revenue'].idxmin(), 'month']

print(f"\nGeneral Trend: Highest sales were in {highest_sales_month}, while lowest sales were in {lowest_sales_month}.")
```

```python
# Visualizing the sales made in each month for trends

x = monthly_sales.month
y = monthly_sales.revenue

#Setting my figure size
plt.figure(figsize=(12,5))

plt.plot(
    x, y, color="green", linestyle="-",
    linewidth=2.0, marker="x", alpha=0.6)

plt.title('Sales made in each month')
plt.xlabel('Months')
plt.ylabel('Revenue Generated')

plt.show()
```

`Observations`: 
- Sales fluctuate throughout the year, with peaks and troughs.
- Revenue generated between `September` and `October` were especially low.
- After `July`, there was a downward spiral till `October`.

### **Question 10**
10.	Are there pizza types that are not doing well on the menu?

```python
# Group by pizza type and sum the quantity
pizza_type_sales = final_merged_df.groupby('pizza_type_id')['quantity'].sum().reset_index(name='total_quantity_sold')

# Sort by total quantity sold in ascending order to find the least popular
least_popular_pizza_types = pizza_type_sales.sort_values(by='total_quantity_sold', ascending=True)

# Display the bottom 5 pizza types
print("\nBottom 5 least popular pizza types:\n")
display(least_popular_pizza_types.head())
```

The pizza type that is not doing well is `brie_carre`. The total quantity sold of `490` is the least

```python
# Visualizing the pizza type that is not doing well on the menu

x = pizza_type_sales.pizza_type_id
y = pizza_type_sales.total_quantity_sold

#Setting my figure size
plt.figure(figsize=(12,5))

plt.plot(
    x, y, color="green", linestyle="-",
    linewidth=2.0, marker="x", alpha=0.6)
plt.xticks(rotation=45)  # Rotate category labels for readability

plt.title('Sales made in each month')
plt.xlabel('Months')
plt.ylabel('Revenue Generated')

plt.show()
```
